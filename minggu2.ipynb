# Langkah 3 - Imputasi

Pada langkah ini kita akan melakukan imputasi terdapat data yang hilang pada "Age", "Cabin", dan "Embarked".

Strategi yang akan kita gunakan adalah,

* "Age" --> Dikarenakan "Age" adalah data nominal, maka kita akan menggunakan strategi mean.
* "Cabin" --> "Cabin" merupakan informasi terkait dengan nomor kabin penumpang. Disini kita akan berasumsi bahwa, seluruh penumpang yang tidak memiliki nomor kabin, merupakan penumpang yang tinggal di dek-dek kapal. Sehingga kita akan mensubtitusi data yang hilang dengan informasi "DECK".
* "Embarked" --> "Embarked" merupakan informasi lokasi embarkasi penumpang. Hanya ada 2 data yang hilang disini. Dikarenakan data merupakan data nominal, maka kita dapat menggunakan modus (mode) untuk mensubtitusi data yang hilang.

```python
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```

# Langkah 0 - Import Library

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
data = '/content/drive/MyDrive/Mesin Learning/wbc.csv'
df6 = pd.read_csv(data)
df6.head()
```

```python
df6.info()
```

# Soal 1

Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
#Nomor 1
df6 = df6[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean','area_worst','symmetry_worst']]
df6.head()
```

# Soal no - 2
Lakukan proses encoding pada kolom "diagnosis"



```python
#Nomor 2
le = LabelEncoder() # membuat objek dari LabelEncoder
df6['diagnosis'] = le.fit_transform(df6['diagnosis']) # proses encoding
df6.head()
```

# Soal no - 3
Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
#Nomor 3
std = StandardScaler() #Inisialisasi Standarisasi

#Proses Standarisasi
df6['radius_mean'] = std.fit_transform(df6[['radius_mean']])
df6['texture_mean'] = std.fit_transform(df6[['texture_mean']])
df6['perimeter_mean'] = std.fit_transform(df6[['perimeter_mean']])
df6['area_mean'] = std.fit_transform(df6[['area_mean']])
df6['smoothness_mean'] = std.fit_transform(df6[['smoothness_mean']])
df6['compactness_mean'] = std.fit_transform(df6[['compactness_mean']])
df6['concavity_mean'] = std.fit_transform(df6[['concavity_mean']])
df6['concave points_mean'] = std.fit_transform(df6[['concave points_mean']])
df6['symmetry_mean'] = std.fit_transform(df6[['symmetry_mean']])
df6['fractal_dimension_mean'] = std.fit_transform(df6[['fractal_dimension_mean']])
df6['area_worst'] = std.fit_transform(df6[['area_worst']])
df6['symmetry_worst'] = std.fit_transform(df6[['symmetry_worst']])
df6.head()
```

# Soal no - 4
Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
#Nomor 4
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:2
df6_train, df6_unseen = train_test_split(df6, test_size=0.2, random_state=0, stratify=df6['diagnosis'])

# Split lagi antara validasi dan testing
df6_val, df6_test = train_test_split(df6_unseen, test_size=0.5, random_state=0, stratify=df6_unseen['diagnosis'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df6.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df6_train.diagnosis.value_counts()}')
print(f'Jumlah label data val:\n{df6_val.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df6_test.diagnosis.value_counts()}')
```
